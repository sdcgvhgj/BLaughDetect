```python
import pandas as pd

df = pd.read_parquet('./AudioSet/data/unbal_train/000.parquet', engine='fastparquet')

print(df[:10])
```

```python
import torch
import torchaudio
import io
import numpy as np
from IPython.display import Audio, display

laugh_idx = []
for idx, row in df.iterrows():
    for label in row['human_labels']:
        if 'laugh' in label or 'Laugh' in label:
            laugh_idx.append(idx)
            break

print(laugh_idx, len(df))

# 选择一行数据
row = df.iloc[laugh_idx[-1]]  # 选择第一行（火车声音）

# 方法1: 使用 BytesIO
audio_bytes = row['audio.bytes']
buffer = io.BytesIO(audio_bytes)

# 用 torchaudio 读取
waveform, sample_rate = torchaudio.load(buffer, format='flac')

print(f"音频形状: {waveform.shape}")
print(f"采样率: {sample_rate} Hz")
print(f"时长: {waveform.shape[1] / sample_rate:.2f} 秒")
print(f"音频内容: {row['human_labels']}")

# 播放音频（在 Jupyter 中）
display(Audio(waveform.numpy(), rate=sample_rate))
```
